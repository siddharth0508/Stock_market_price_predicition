```python
import numpy as np
import pandas as pd
from pandas_datareader import data, wb
import datetime
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
from pandas import datetime
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import warnings
import plotly.express as px
```

```python
to_date=datetime.now()
start_date = to_date - relativedelta(years=5)
dates = pd.date_range(start_date, to_date)
df = data.DataReader('DELL', 'yahoo', dates[0],dates[-1],)
#df['Date']=df.index
#df.reset_index(inplace=True, drop=True)
df
```

```python
df1=df[['Close']]
df1
```

```python
forecast_out=7
df1['prediction']=df1[['Close']].shift(-forecast_out)
df1
```

```python
X=np.array(df1.drop(['prediction'],1))
X=X[:-forecast_out]
X
```

```python
Y=np.array(df1['prediction'])
Y=Y[:-forecast_out]
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
svr_rbf=SVR(kernel='rbf',C=1e3,gamma=0.1)
svr_rbf.fit(x_train,y_train)
```

```python
svm_confidence=svr_rbf.score(x_test,y_test)
svm_confidence
```
